```python
class SongNode:
    def __init__(self, title, artist):
        self.title = title
        self.artist = artist
        self.next = None

class SpotifyLite:
    def __init__(self):
        self.head = None

    def add_song(self, title, artist):
        new_song = SongNode(title, artist)
        if not self.head:
            self.head = new_song
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_song
        print(f"‚úÖ Added: {title} by {artist}")

    def display_playlist(self):
        if not self.head:
            print("üéµ Playlist is empty.")
            return
        current = self.head
        print("\nüéß Your Playlist:")
        while current:
            print(f"üéµ {current.title} - {current.artist}")
            current = current.next

    def sort_playlist(self):
        if not self.head or not self.head.next:
            return
        swapped = True
        while swapped:
            swapped = False
            current = self.head
            while current.next:
                if current.title.lower() > current.next.title.lower():
                    current.title, current.next.title = current.next.title, current.title
                    current.artist, current.next.artist = current.next.artist, current.artist
                    swapped = True
                current = current.next
        print("\n‚úÖ Playlist sorted by song title!")

    def search_song(self, title):
        current = self.head
        found = False
        while current:
            if current.title.lower() == title.lower():
                print(f"üîç Found: {current.title} by {current.artist}")
                found = True
                break
            current = current.next
        if not found:
            print(f"‚ùå Song '{title}' not found in playlist.")

    def delete_song(self, title):
        current = self.head
        prev = None
        while current:
            if current.title.lower() == title.lower():
                if prev:
                    prev.next = current.next
                else:
                    self.head = current.next
                print(f"üóëÔ∏è Deleted: {current.title} by {current.artist}")
                return
            prev = current
            current = current.next
        print(f"‚ùå Song '{title}' not found. Cannot delete.")

spotify = SpotifyLite()
spotify.add_song("Believer", "Imagine Dragons")
spotify.add_song("Shape of You", "Ed Sheeran")
spotify.add_song("Bohemian Rhapsody", "Queen")
spotify.add_song("Levitating", "Dua Lipa")
spotify.display_playlist()
spotify.search_song("Shape of You")
spotify.search_song("Despacito")
spotify.delete_song("Shape of You")
spotify.delete_song("Despacito")
spotify.display_playlist()
spotify.sort_playlist()
spotify.display_playlist()
```
